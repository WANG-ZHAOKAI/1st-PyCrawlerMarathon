### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

```python
for a in soup.find(class_="part_list_2").find_all('h3'):
    print( a.find_all('span')[0].text, a.find_all('em')[0].text, a.find_all('a')[0].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
cate = {}
for a in soup.find(class_="part_list_2").find_all('h3'):
    if not a.find_all('em')[0].text in cate:
        cate_name = a.find_all('em')[0].text
        cate.setdefault(cate_name)
        cate[cate_name] = []
        cate[cate_name].append(a.find_all('a')[0].text)
    else:
        cate_name = a.find_all('em')[0].text
        cate[cate_name].append(a.find_all('a')[0].text)
print(cate)
```

### ③ 哪一個類別的文章最多

```python
for i in cate:
    print(i,':',len(cate[i]))

        
```
