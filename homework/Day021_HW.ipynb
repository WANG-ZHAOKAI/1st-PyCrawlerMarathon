## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

## 2. 取出現在時間兩小時內的新聞

```python
import datetime

for d in soup.find(class_="part_list_2").find_all('h3'):
    time = datetime.datetime.strptime(d.find(class_="date").text,"%Y/%m/%d %H:%M")
    if time.hour >= 13 and time.hour <= 15:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
browser.quit();
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
day = datetime.date.today()-datetime.timedelta(days=3)
url = 'https://www.ettoday.net/news/news-list-{}-{}-{}-0.htm'.format(day.year, day.month, day.day)
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(url)
```

```python
import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    time = datetime.datetime.strptime(d.find(class_="date").text,"%Y/%m/%d %H:%M")
    if time.hour >= 14 and time.hour <= 17:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
browser.quit();
```
