### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")
```

```python
title = []
for d in soup.find_all(class_="title"): 
    try:
        soup_2 =BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        author=[]
        for d2 in soup_2.find_all(class_="article-meta-value"):
            author.append(d2.text)
        title.append(author)
    except:
        continue
df = pd.DataFrame(title,columns=['作者','看板','標題','時間'])
df
```
